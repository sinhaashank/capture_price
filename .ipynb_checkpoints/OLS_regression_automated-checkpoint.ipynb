```python
import pandas as pd
import numpy as np
import os
import sys
import plotly
import plotly.graph_objects as go
```

```python
#### Importing the Dataset
dataset = pd.read_excel('ES_solar_m.xlsx')
```

```python
dataset = dataset.set_index('Timestamp')
```

```python
dataset = dataset.dropna()
```

```python
dataset
```

```python
col_names = dataset.columns[1:].to_list()
```

```python
X = dataset.iloc[:,1:].values
y = dataset.iloc[:,0].values
```

```python
import statsmodels.api as sm

def backwardElimination(X, SL, col_names):
    numVars = len(X[0])
    len_data = len(X[:,0])
    X = np.append(arr = np.ones((len_data, 1)).astype(int), values = X, axis = 1)
    temp = np.zeros((len_data,1))
    for i in range(0, numVars):
        regressor_OLS = sm.OLS(y, X).fit()
        maxVar = max(regressor_OLS.pvalues[1:]).astype(float)
        adjR = regressor_OLS.rsquared_adj.astype(float)
        if maxVar > SL:
            for j in range(1, numVars + 1 - i):
                if (regressor_OLS.pvalues[j].astype(float) == maxVar):
                    temp = X[:, j]
                    tempX = X
                    X = np.delete(X, j, 1)
                    tmp_regressor = sm.OLS(y, X).fit()
                    adjR_new = tmp_regressor.rsquared_adj.astype(float)
                    if (adjR > adjR_new):
                        return (tempX,col_names)
                    else:
                        print('Removing ' + str(col_names[j]))
                        col_names.pop(j)
                        break
    return (X,col_names)
```

```python
SL = 0.05
X_opt = X
col_names = dataset.columns[1:].to_list()
col_names.insert(0,'Intercept')

[X_Modeled, x_sig] = backwardElimination(X_opt, SL, col_names)

ols = sm.OLS(y, X_Modeled).fit()

print (x_sig)
ols.summary(xname = x_sig)
#ols.summary()
```

```python
y_pred=ols.predict(X_Modeled)
```

```python
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=dataset.index, y=y,
    name='real',
    mode='lines',
))

fig.add_trace(go.Scatter(
    x=dataset.index, y=y_pred,
    name='predicted',
    mode = 'markers'    
))

fig.update_layout(title='Solar Capture Price in Spain')

fig.update_yaxes(title='Capture Price (€/MWh)')

fig.show()
```
