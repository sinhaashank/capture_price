```python
X = dataset.iloc[:,2:].values
y = dataset.iloc[:,1].values
```

```python
import statsmodels.api as sm

def backwardElimination(X, SL, col_names):
    numVars = len(X[0])
    X = np.append(arr = np.ones((60, 1)).astype(int), values = X, axis = 1)
    temp = np.zeros((60,1))
    for i in range(0, numVars):
        regressor_OLS = sm.OLS(y, X).fit()
        maxVar = max(regressor_OLS.pvalues[1:]).astype(float)
        adjR = regressor_OLS.rsquared_adj.astype(float)
        if maxVar > SL:
            for j in range(0, numVars - i):
                if (regressor_OLS.pvalues[j+1].astype(float) == maxVar):
                    temp = X[:, j+1]
                    tempX = X
                    X = np.delete(X, j+1, 1)
                    tmp_regressor = sm.OLS(y, X).fit()
                    adjR_new = tmp_regressor.rsquared_adj.astype(float)
                    if (adjR > adjR_new):
                        #X_rollback = np.hstack((X, temp))
                        #X_rollback = np.append(arr = np.ones((60, 1)).astype(int), values = X_rollback, axis = 1)
                        #X_rollback = np.delete(X_rollback, j, 1)
                        return (tempX,col_names)
                    else:
                        print('Removing ' + str(col_names[j]))
                        col_names.pop(j)
                        break
    return (tempX,col_names)
```

```python
SL = 0.05
X_opt = X

[X_Modeled, col_names] = backwardElimination(X_opt, SL, col_names)

ols = sm.OLS(y, X_Modeled).fit()
col_names.insert(0,'Intercept')

ols.summary(xname = col_names)
```
